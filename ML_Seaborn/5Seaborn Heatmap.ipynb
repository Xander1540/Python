# Seaborn Heatmap()

graphical replresentation of 2D data.

```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr= np.linspace(1,5, 12).reshape(4,3)
arr
```

```python
sns.heatmap(arr)
```

```python
df= pd.read_csv("Who_is_responsible_for_global_warming.csv")
df
```

```python
x= df.corr()
print(x)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(x, annot=True)
```
