# Matplotlib Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age= np.random.randint(18,45, (100))
py_students_age= np.random.randint(15,40, (100))
display(ml_students_age, py_students_age)
```

```python
bins= [15,20,25,30,35,40,45] #to set range of bar on x axis
plt.hist(ml_students_age, bins, rwidth=0.8)# rwidth
plt.title("ML students age histogram")
plt.xlabel("ML students age")
plt.ylabel("No. of students")
plt.show()
```

```python
# histtype= "step, bar, barstacked, stepfilled"
# orientation= 'horizontal, vetical'

bins= [15,20,25,30,35,40,45] #to set range of bar on x axis
plt.figure(figsize= (16,9)) # resize graph 
plt.hist(ml_students_age, bins, rwidth=0.8, histtype='bar', orientation='horizontal',
         color='y')
plt.hist(py_students_age, bins, rwidth=0.8, histtype='bar', orientation='vertical',
         color='b')
plt.title("ML students age histogram")
plt.xlabel("ML students age")
plt.ylabel("No. of students")
plt.legend("ML Student", loc=4)
plt.show()
```

```python
bins= [15,20,25,30,35,40,45] #to set range of bar on x axis
plt.figure(figsize= (16,9)) # resize graph 
plt.hist(ml_students_age, bins, rwidth=0.8, histtype='bar', orientation='horizontal',
         color='y')
plt.hist(py_students_age, bins, rwidth=0.8, histtype='bar', orientation='vertical',
         color='b')
plt.grid(color="k", linestyle="-", linewidth=1)
plt.title("ML students age histogram")
plt.xlabel("ML students age")
plt.ylabel("No. of students")
plt.legend("ML Student", loc=4)
plt.show()
```
