# Matplotlib BarChart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes, class1_students)
plt.show()
```

```python
plt.barh(classes, class1_students)
plt.show()
```

alpha is used to change opacity of bar

```python
plt.bar(classes, class1_students, width=0.2, align='edge', color='y', edgecolor='black', 
        linewidth=3, alpha=0.9, linestyle='--')
plt.show()
```

```python
plt.grid()
plt.bar(classes, class1_students, width=0.4, align='edge', color='y', edgecolor='black', 
        alpha=0.9, linestyle='-')
plt.title("Class", fontsize=15)
plt.xlabel("Subjects")
plt.ylabel("No. of students")
plt.legend("ML Student", loc=1)
plt.show()
```

```python
plt.grid()

class_index= np.arange(len(classes))
width=0.2
plt.bar(classes, class1_students, width, align='edge', color='y', edgecolor='black')
plt.bar(class_index + width, class2_students, width, align='edge', color='r', edgecolor='black')
plt.bar(class_index + width + width, class3_students, width, align='edge', color='g', edgecolor='black')
plt.title("Class", fontsize=15)
plt.xlabel("Subjects")
plt.ylabel("No. of students")
plt.show()
```
