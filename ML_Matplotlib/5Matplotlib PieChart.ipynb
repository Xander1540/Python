# Matplotlib PieChart

```python
import matplotlib.pyplot as plt
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10]
plt.pie(class1_students)
plt.show()
```

`autopct` is used for marking percentage
<br> to slice and show a particular data from the chart `explode`
<br> `shadow=True` is used for showing shadow 


```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 15, 20, 25, 10]
ex=[0,0,0.1,0,0]
colors= ['c', 'b', 'r', 'y', 'g']

s ={"fontsize": 20}

plt.pie(class1_students, labels=classes, explode=ex, colors=colors,
        autopct="%0.2f%%", radius=1.5, shadow=True, textprops=s)
plt.show()
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 15, 20, 25, 10]
ex=[0,0,0.3,0,0]
colors= ['c', 'b', 'r', 'y', 'g']

s ={"fontsize": 10}

plt.pie(class1_students, labels=classes, explode=ex, colors=colors,
        autopct="%0.2f%%", radius=5.5, shadow=True, textprops=s, frame=True)
plt.legend(loc=2)
plt.show()
```
