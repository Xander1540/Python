# Pandas fillna()

```python
import pandas as pd
```

```python
f= pd.read_csv('C:\\PythonPractice\\Fortune_5.csv')
f
```

```python
f.fillna(0)
```

```python
f.fillna({'Name': 'none', 'Profit': 0, 'Growth': 0})
```

```python
# fills the NaN value with previous value of col.
f.fillna(method= 'ffill')
```

```python
# same as 'bfill'
f.fillna(method= 'pad')
```

```python
# fills the NaN value with next value of col.
f.fillna(method='bfill')
```

```python
# by default axis is col.
f.fillna(method= 'ffill', axis= 0)
```

```python
print(f)
```

```python
# axis =1 , NaN value is filled by value of row.
f.fillna(method= 'ffill', axis= 1)
```

```python
# limit set limit to no. of NaN value that'll change.
f.fillna(0, limit= 1) 
```

```python
# inplace= True
# edits the original data frame
f.fillna(5, inplace=True)
f
```
