# Pandas concat()

```python
import pandas as pd
```

```python
sr1= pd.Series([1,2,3])
sr1
```

```python
sr2= pd.Series([4,5,6])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
df1= pd.DataFrame({'ID': [1,2,3,4], 'Name': ['A', 'B', 'C', 'D'], 'Class': [5,6,7,8]})
df1
```

```python
df2= pd.DataFrame({'ID': [5,6,7,8], 'Name': ['E', 'F', 'G', 'H'], 'Class': [9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
# if we want to concatinate dataframes column-wise (side-by-side)
pd.concat([df1,df2], axis=1)
```

```python
# if we want to change index and make it acc. to new dataframe (concatinated)
pd.concat([df1,df2], ignore_index=True)
```

```python
df1= pd.DataFrame({'ID': [1,2,3,4], 'Name': ['A', 'B', 'C', 'D'], 'Class': [5,6,7,8]})
df1
```

```python
df4= pd.DataFrame({'ID': [3,4], 'Name': ['C', 'D'], 'Class': [7,8]})
df4
```

```python
pd.concat([df1,df4])
```

```python
pd.concat([df1,df4], axis=1)
```

```python
# if we want to print only common value of 2 dataframes
pd.concat([df1,df4], axis=1, join='inner')
```

```python
pd.concat([df1,df2], axis=1, join='outer')
```

```python
df2= pd.DataFrame({'ID': [5,6,7,8], 'Name': ['E', 'F', 'G', 'H'], 'Class': [9,10,11,12]})
df2
```

```python
# giving label to indexes of dataframes concatinated 
pd.concat([df1,df2], keys=['DF1','DF2'])
```

```python
pd.concat([df1,df2], axis=1, keys=['DF1','DF2'])
```

```python
df5= pd.DataFrame({'Marks': [80,90,100,100]})
df5
```

```python
pd.concat([df1,df5])
```
